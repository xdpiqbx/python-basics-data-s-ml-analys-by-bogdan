```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
values = np.array([0, 1, 2, 3, 4, 5, 6, 7])
squares = values ** 2
squares
```

```python
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values = np.array(range(100))
pows = values ** 3
```

```python
plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=10)
plt.show()
```

```python
values = np.array(list(range(10)))
squares = values ** 2
cubes = values ** 3
```

```python
figure, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.scatter(values, squares)
ax1.set_title('Square nums')
ax2.plot(values, cubes)
ax2.set_title('Cube nums')
plt.show()
```

```python
figure, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.scatter(values, squares)
ax1.set_title('Square nums')
ax2.plot(values, cubes)
ax2.set_title('Cube nums')
plt.show()
```

```python
data = {
    'Year': np.array(range(2001, 2007)),
    'Sales': np.random.randint(100, 270, 6)
}
data
```

```python
data = {
    'Year': np.array(range(2001, 2007)),
    'Sales': np.random.randint(100, 270, 6)
}
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='line')
plt.show()
```

```python
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='bar')
plt.show()
```

```python
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind='area')
plt.show()
```

```python
df = pd.DataFrame(data)
df.plot(y='Sales', kind='pie', labels=df['Year'], autopct='%1.1f%%')
plt.show()
```

```python
data = {
    'Age': np.random.randint(25, 60, 100),
}
df = pd.DataFrame(data)
df.plot(y='Age', kind='hist')
plt.title('Age distribution')
plt.show()
```

```python
categories = ['FIRST', 'SECOND', 'THIRD']
data = {
    'Category': np.random.choice(categories, 10),
    'Value': np.random.choice(np.arange(0, 100, 5), 10)
}
df = pd.DataFrame(data)
df.head()
```

```python
df.boxplot(by='Category', column='Value')
plt.title('Values distribution by Category')
plt.show()
```

```python
monthes = ['January', 'February', 'March', 'April', 'May']
data = {
    'Month': monthes,
    'Week 1 Temp': np.random.choice(np.arange(4, 25), len(monthes)),
    'Week 2 Temp': np.random.choice(np.arange(4, 25), len(monthes)),
    'Week 3 Temp': np.random.choice(np.arange(4, 25), len(monthes)),
    'Week 4 Temp': np.random.choice(np.arange(4, 25), len(monthes)),
}
df = pd.DataFrame(data)
df.head()
```

```python
df.set_index('Month', inplace=True)
df.head()
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly temperature data')
```
